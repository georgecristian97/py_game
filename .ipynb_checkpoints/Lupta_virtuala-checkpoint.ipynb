```python
class Item:
    def __init__(self, arma, dmg,crit):
        self.arma = arma
        self.dmg = dmg
        self.crit = crit
```

```python
import time
```

```python
from pygame import mixer  # Load the popular external library
mixer.init()
mixer.music.load('Rammstein - Mein Herz Brennt (Instrumental) (432Hz) (mp3cut.net).mp3')
mixer.music.play(-1) # 5 de 5 ori|||| -1 infty 
mixer.music.set_volume(0.25)
```

```python
class Luptator(Item):
    def __init__(self, nume, hp,apar):
        self.nume = nume
        self.hp = hp
        self.apar = apar
```

```python
class Batalie(Luptator):
    def __init__(self,bat):
        self.bat=bat
    def lupta(self,L1,L2):
        turn =0;
        print(L1.nume," vs. ",L2.nume)
        print("Figth! ",turn)
                # basic example for multi-line printing 
        print( 
        ''' 
======================================= 
|                                     |
|        >()       \  ()              | 
|     ,_ /||\_,     \/||\             | 
|         /\          /\              | 
======================================= 
        '''
            )     
        time.sleep(1)
        self.start_(L1,L2,turn)
    def attack(self,L1,L2,turn):
        time.sleep(0.55)
        
        if (L2.hp -(L1.dmg*(L1.crit/100))+L2.apar > L2.hp):
            print(turn," : ",L1.nume," miss ",L2.nume)
            print( 
            ''' 
======================================= 
|                                     |
|         >()\ ()                     | 
|         /\\/\//\                    | 
|        /\      /\                   | 
======================================= 
            '''
                ) 
        elif (L2.hp - (L1.dmg * (L1.crit / 100)) + L2.apar == L2.hp):
            print(turn," : ",L1.nume," block ",L2.nume)
            print( 
             ''' 
======================================= 
|                                     |
|         >()\ ()                     | 
|         /\\/\//\                    | 
|        /\      /\                   | 
======================================= 
             '''
                ) 
        elif (L2.hp - (L1.dmg * (L1.crit / 100)) + L2.apar <= 0):
            L2.hp = L2.hp - (L1.dmg * (L1.crit / 100)) + L2.apar;
            print(turn," : ",L1.nume," killed ",L2.nume,": ",L2.hp, " hp")
            if (L1.nume =="Ninja"):
                print( 
                ''' 
======================================= 
|                                     |
|         >()/                        | 
|       ,_/||                         | 
|          /\    ==>o                 | 
======================================= 
                '''
                ) 
            else:
                print( 
                ''' 
======================================= 
|        /                            |
|       /                             |
|       \()/                          |
|        ||                           | 
|        /\    ==>o                   | 
======================================= 
                '''
                ) 
        else:
            print(turn," : ",L1.nume," hit ",L2.nume," : - ",(L1.dmg * (L1.crit / 100)) - L2.apar," hp")
            L2.hp = L2.hp - (L1.dmg * (L1.crit / 100)) + L2.apar
            if (L1.nume =="Ninja"):       
                print( 
                ''' 
======================================= 
|                                     |
|         >()  / ()                   | 
|         /\\_,/-||-                   | 
|        /\    /\                     | 
======================================= 
                '''
                ) 
            else:
                print( 
                ''' 
======================================= 
|                                     |
|                                     |
|         >()     ()                  |
|          -||-___//\                 | 
|           /\      /\                | 
======================================= 
                '''
                ) 
                
        print(L1.nume," : ",L1.hp," hp | ",L2.nume ," : ", L2.hp," hp")
        return L2.hp
        
    def start_(self, L1,L2,turn):
        while((L1.hp > 0) and (L2.hp>0) and (L1.apar>0)):
            turn = turn +1  
            L2.hp=self.attack(L1,L2,turn)
            if (L2.hp>0): 
                L1.hp = self.attack(L2,L1,turn)
            L2.apar=L2.apar-3.4567
            L1.apar=L1.apar-3.4567
            #end if (L2.hp>0): 
        #end while((L1.hp > 0) and (L2.hp>0)): 
```

```python
m1 = Luptator("Razboinic",2000,400)

m1.dmg = 180
m1.crit= 200
m1.arma= "Sword"

m2 = Luptator("Ninja",1800,350)

m2.dmg = 230
m2.crit= 210
m2.arma= "Dagger"
```

```python
f = Batalie("x")

f.lupta(m1,m2)
```
